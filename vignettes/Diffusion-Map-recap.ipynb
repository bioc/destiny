```r
library(destiny)
data(guo)
```

It can be used directly for diffusion map creation:

```r
dm_guo <- DiffusionMap(guo, verbose = FALSE,
                       censor_val = 15, censor_range = c(15, 40))
dm_guo
```

```r
plot(dm_guo, pch = 20)
```
